## **Daily Expense Tracker using Python and NumPy**

**Problem Statement**

The objective of this project is to build a Python program that allows the user to enter the daily expenses for multiple days and store them using NumPy arrays, and then analyze them. The program calculates the total expense, average expense, and the highest expense, and finally checks whether the user's average spending exceeds a specified limit.

**Process Explanation**

The input is taken by asking the user how many days of expenses they want to record.
Using a loop, all the expenses are collected and stored in a List.
This list is then converted into a NumPy array for easier numerical processing.
NumPy functions are used to calculate the total expense, the average spending, and to identify the highest spending day.
A spending summary is then displayed to the user showing all calculated values.
Finally, a warning message is shown if the user's average daily expense crosses a predefined spending limit.

```python
import numpy as np

# User-defined function for warning
def expense_limit(avg_expense , limit):
    if avg_expense > limit:
      return "Warning: Your average expense exceeds the limit!"
    else:
      return "Your average expense is within the limit!"

# Input: number of days
days = int(input("Enter number of days: "))

# Collecting Daily Expenses Through Loops
expenses = []
for i in range(days):
  amount = float(input(f"Enter expense for day {i+1}: "))
  expenses.append(amount)

# Converting List to NumPy Array
expenses_ = np.array(expenses)

total_expense = np.sum(expenses_)
avg_expense = np.mean(expenses_)
highest_expense = np.max(expenses_)

# Setting a spending limit as per the User
limit = float(input("Enter your average spending limit: "))

# Checking the limit using functions set above
limit_ = expense_limit(avg_expense, limit)

print("SUMMARY OF EXPENSES")
print(f"Total Expense  :  {total_expense}")
print(f"Average Expense:  {avg_expense}")
print(f"Highest Expense: {highest_expense}")
print(f"Limit Status  : {limit_}")
```
