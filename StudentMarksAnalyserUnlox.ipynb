### **Student Marks Analyser Using Python and NumPy**

**Problem Statement**

The Objective of this Project is to build a Python Program that allows the user to take the Marks of multiple students and store them using NumPy Arrays and then analysing them. The program calculates the total marks, average marks, the highest and lowest marks, and finally assign a grade based on the average score of the students.

**Process Explanation**

1. The Input is taken consisting of how many student's marks are to be entered.
2. Using a loop, all the marks are collected and stored in a List.
3. This list is then converted into a NumPy array.
4. NumPy library is used to calculate the sum, mean, highest, and lowest values.
5. In the end, a grade is assigned using conditional statements based on the average marks.

Grading Pattern:

• A+ : 90 and above

• A : 80–89

• B : 70–79

• C : 60–69

• D : 50–59

• Fail : below 50

```python
import numpy as np

# User-defined function to calculate grade from average
def grades_students(average_marks) :
  if average_marks >= 90:
    return "A+"
  elif average_marks >= 80:
    return "A"
  elif average_marks >= 70:
    return "B"
  elif average_marks >= 60:
    return "C"
  elif average_marks >= 50:
    return "D"
  else:
    return "Fail"

# Input: number of students
no_of_students = int(input("Enter number of students: "))

# Collection of marks using a loop and a list
marks = []
for i in range (no_of_students):
  score = float(input(f"Enter marks of student {i+1}: "))
  marks.append(score)

# Converting Python list into a NumPy array
marks_ = np.array(marks)

total_marks = np.sum(marks_)
average_marks = np.mean(marks_)
highest_marks = np.max(marks_)
lowest_marks = np.min(marks_)

# Using the user-defined function for grading
final_grade = grades_students(average_marks)

print("STUDENT MARKS ANALYSIS")
print(f"Marks: {marks_}")
print(f"Total Marks = {total_marks}")
print(f"Average Marks = {average_marks}")
print(f"Highest Marks = {highest_marks}")
print(f"Lowest Marks = {lowest_marks}")
print(f"Assigned Grade = {final_grade}")
```
